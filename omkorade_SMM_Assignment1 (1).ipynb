```python
# Function to classify if a given comment is toxic or not by adding a new column of True/False based on maximum True/False values from composite_toxic column
def classify_toxicity(composite_toxic):
  bool_values = [item[0] for item in composite_toxic]
  return sum(bool_values) >= 3

training_df['Result'] = training_df['composite_toxic'].apply(classify_toxicity)
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
stop_words = stopwords.words('english')

lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    if isinstance(text, str):
    # Lowercasing
        text = text.lower()
    # Remove punctuation
        text = re.sub(r'[^\w\s]', '', text)
    # Tokenization
        tokens = text.split()
    # Remove stop words
        tokens = [token for token in tokens if token not in stop_words]
    # Lemmatization
        tokens = [lemmatizer.lemmatize(token) for token in tokens]
    # Join tokens back into a string
        return ' '.join(tokens)
    else:
        return ""

# preprocessing of the 'text' columns
training_df['text_processed'] = training_df['text'].apply(preprocess_text)
test_df['text_processed'] = test_df['text'].apply(preprocess_text)
#df['title_processed'] = df['article_title'].apply(preprocess_text)

print(training_df['text_processed'].head())
print(test_df['text_processed'].head())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.utils.class_weight import compute_class_weight
```

```python
# Feature extraction using TF-IDF
vectorizer = TfidfVectorizer(max_features = 5000, ngram_range=(1,2))

X_train = vectorizer.fit_transform(training_df['text_processed'])
X_test = vectorizer.transform(test_df['text_processed'])

y_train = training_df['Result']  # Target variable

class_weights = compute_class_weight(class_weight="balanced",
                                     classes=np.array([0, 1]),
                                     y=training_df['Result'])

# Train the Logistisc Regression model
model = LogisticRegression(class_weight={0: class_weights[0], 1: class_weights[1]}, max_iter=1000)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(y_pred)

```

```python
# Train the SVM model
model = SVC(class_weight={0: class_weights[0], 1: class_weights[1]})
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(y_pred)

```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
```

```python
training_df['Result'] = training_df['Result'].astype(int)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_df, val_df = train_test_split(training_df, test_size=0.2, stratify=training_df['Result'], random_state=42)
```

```python
import torch
```

```python
class ToxicDataset(torch.utils.data.Dataset):
    def __init__(self, dataframe, tokenizer):
        self.dataframe = dataframe
        self.tokenizer = tokenizer
        self.texts = dataframe['text'].tolist()
        self.labels = dataframe['Result'].tolist()

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = self.texts[idx]
        label = self.labels[idx]
        encoding = self.tokenizer(text, truncation=True, padding='max_length', max_length=128, return_tensors='pt')  # padding='max_length'

        return {
            'input_ids': encoding['input_ids'].flatten(), # Flatten
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': torch.tensor(label)
        }
```

```python
train_dataset = ToxicDataset(train_df, tokenizer)
val_dataset = ToxicDataset(val_df, tokenizer)
#test_dataset = ToxicDataset(test_df, tokenizer)
```

```python
from transformers import DataCollatorWithPadding

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer

model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=2)
```

```python
from transformers import EarlyStoppingCallback
```

```python
training_args = TrainingArguments(
    output_dir="./results",
    learning_rate=1e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=5,  # epochs
    weight_decay=0.1,
    evaluation_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    metric_for_best_model="eval_loss", # "eval_accuracy"
    report_to="none", # Remove wandb
    logging_steps=50
)

# Define Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset, # validation set
    tokenizer=tokenizer,
    data_collator=data_collator,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=2)]
)

trainer.train()
```

```python
class ToxicTestDataset(torch.utils.data.Dataset):
    def __init__(self, dataframe, tokenizer):
        self.texts = dataframe['text'].tolist()
        self.tokenizer = tokenizer

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = self.texts[idx]
        encoding = self.tokenizer(text, truncation=True, padding='max_length', max_length=128, return_tensors='pt')

        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten()
        }

# Create test dataset
test_dataset = ToxicTestDataset(test_df, tokenizer)
```

```python
from torch.utils.data import DataLoader

# Define DataLoader
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)

predictions = []

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

with torch.no_grad():
    for batch in test_loader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        logits = outputs.logits
        preds = torch.argmax(logits, dim=1).cpu().numpy()
        predictions.extend(preds)

boolean_predictions = [bool(pred) for pred in predictions]

test_df['Predicted_Label'] = boolean_predictions

print(test_df[['platform_id', 'Predicted_Label']])
```
